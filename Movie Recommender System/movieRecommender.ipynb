# **importiing the dependicies**


```python
import numpy as np
import pandas as pd
import difflib  #to find the name of the movie closest to the name given by user
from sklearn.feature_extraction.text import TfidfVectorizer   #convert text data into feature vectors
from sklearn.metrics.pairwise import cosine_similarity # to find the item which is most similar
```

# **Data** **Preprocessing**

```python
movies = pd.read_csv('/content/movies.csv')
```

```python
#printing first five rows
movies.head()
```

```python
print(movies.shape)
movies.columns
```

```python
#selecting relevent columns/features
selected_features = ['genres','keywords', 'production_companies','cast', 'director', 'tagline']
print(selected_features)
```

```python
# replacing null values with null string
for feature in selected_features:
  movies[feature] = movies[feature].fillna('')
```

```python
# combining all selected features
combined_features = movies['genres']+''+movies['keywords']+''+movies['production_companies']+''+movies['cast']+''+movies['director']+''+movies['tagline']
print(combined_features)
```

```python
# converting the text data into numerical data
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

# **Cosine Similirity**

```python
similirity = cosine_similarity(feature_vectors)
```

```python
print(similirity)
```

```python
similirity.shape
```

```python
movie_name = input("enter your favourite movie name-")
```

```python
# creating the list of simillar movies from the dataset
movie_titles = movies['title'].tolist()
print(movie_titles)
```

```python
#finding the close match
close_match = difflib.get_close_matches(movie_name , movie_titles)
print(close_match)
```

```python
closest_match = close_match[0]
print(closest_match)
```

```python
#find the index of the movie
index_of_movie = movies[movies.title == closest_match]['index'].values[0]
print(index_of_movie)
```

```python
similirity_score = list(enumerate(similirity[index_of_movie]))
print(similirity_score)
len(similirity_score)
```

```python
similirity_score = sorted(similirity_score , key = lambda x:x[1], reverse = True)
print(similirity_score)
```

```python
#printing the names of similar movies
movies = pd.read_csv('/content/movies.csv')
print('Movies suggested for you: \n')
i = 1
for movie in similirity_score:
  index = movie[0]
  title_from_index = movies[movies.index == index]['title'].values[0]
  if i<11:
    print(i,":",title_from_index,"\n")
    i+=1
```

# **Final Recommendation System**

```python
movie_name = input("enter your favourite movie name-")
movie_titles = movies['title'].tolist()
close_match = difflib.get_close_matches(movie_name , movie_titles)
closest_match = close_match[0]
index_of_movie = movies[movies.title == closest_match]['index'].values[0]
similirity_score = list(enumerate(similirity[index_of_movie]))
similirity_score = sorted(similirity_score , key = lambda x:x[1], reverse = True)

movies = pd.read_csv('/content/movies.csv')
print('Movies suggested for you: \n')
i = 1
for movie in similirity_score:
  index = movie[0]
  title_from_index = movies[movies.index == index]['title'].values[0]
  if i<11:
    print(i,":",title_from_index,"\n")
    i+=1
```
